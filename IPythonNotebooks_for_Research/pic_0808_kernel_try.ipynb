```python
def my_kernel_eu_weighted(x, y, gamma=.1):
    global myweight
    x = x * myweight
    y = y * myweight
    f = pairwise.euclidean_distances
    dis = f(x, y, squared=True)
    dis /= 80.
    dis *= -gamma
    np.exp(dis, dis)
    return dis
```

```python
def my_kernel_eu(x, y, gamma=.1):
    f = pairwise.euclidean_distances
    dis = f(x, y, squared=True)
    dis /= 80.
    dis *= -gamma
    np.exp(dis, dis)
    return dis
```

```python
def my_kernel_mh_squared(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis = dis**2
    dis *= -gamma
    np.exp(dis, dis)
    return dis
```

```python
def my_kernel_mh(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis *= -gamma
    np.exp(dis, dis)
    return dis
```

```python
def my_kernel_linear(x, y, gamma=.1):
    """gamma stands for the slope"""
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    #------ inear kernel, inplace calculate
    dis *= -gamma
    dis += 1.
    #dis[dis < 0] = 0.
    return dis
```

```python
def my_kernel_rquad(x, y, gamma=.1):
    """gamma stands for the constant C"""
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    c = gamma
    dis += c
    return c/dis
```

```python
def my_kernel_rquad2(x, y, gamma=.1):
    """gamma stands for the constant C"""
    f = pairwise.euclidean_distances
    dis = f(x, y, squared=True)
    dis /= 80.
    c = gamma
    dis += c
    return c/dis
```

```python
def my_kernel_spherical(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis /= 4.
    dis[dis > 1.] = 1.
    f_s = lambda x: 1 - x * 3 / 2. + x**3 / 2.
    return f_s(dis)
```

```python
def my_kernel_spherical2(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    #--------- transformation
#     cut_lim = 4.
#     median = np.median(dis)
#     rate = (2. - 1.)/(cut_lim - median)
#     translation = 1 - rate * median
    #--------- transformation
#     dis *= rate
#     dis += translation
    dis[dis > 6.] = 6.
    f_s = lambda x: 0.5 - (x/3.-1) * 3/4. + (x/3-1)**3 / 4.
    return f_s(dis)
```

```python
def my_kernel_circular(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis /= 4.
    dis[dis > 1.] = 1.
    f_s = lambda x: np.arccos(-x) - x * np.sqrt(1 - x**2)
    return np.pi / f_s(dis) - 1
```

```python
def my_kernel_circular2(x, y, gamma=.1):
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis /= 2.
    dis[dis > 2.] = 2.
    f_s = lambda x: 1 - (x-1) * 3 / 2. + (x-1)**3 / 2.
    return f_s(dis)
```

```python
def my_kernel_power(x, y, gamma=.1):
    """gamma stands for transformation rate"""
    f = pairwise.manhattan_distances
    dis = f(x, y)
    dis /= 80.
    dis *= gamma
    dis += 1.
    return np.power(dis, -2)
```

```python
def my_kernel_dot(x, y, gamma=.1):
    dot_product = np.dot(x, y.T)
    dot_product *= gamma
#     sign = -np.sign(dot_product)
#     dot_product *= sign
    # that is: np.sign(dot_product) * np.exp(np.abs(dot_product))
    return dot_product
```
