```python
def MyRgrs(x, yin, yout, model=None, norm=False):
    '''use yin and yout 's index to align. x is of full length.

    Parameters
    ----------
    model : sklearn clf

    Returns
    -------
    res : regression result
    
    '''
    Xin = x.ix[yin.index]
    Xout = x.ix[yout.index]
    if not Xin.ndim > 1:
        Xin = Xin.reshape(-1, 1)
        Xout = Xout.reshape(-1, 1)
    if norm:
        Xin = preprocessing.normalize(Xin, norm='l2')
    res = model.fit(Xin, yin)
    rsq_in = res.score(Xin, yin)
    print ('rsq_in: %f' % rsq_in)
    rsq_out = res.score(Xout, yout)
    print ('rsq_out: %f' % rsq_out)
    return res
```
